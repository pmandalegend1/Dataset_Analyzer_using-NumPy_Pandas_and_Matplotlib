```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
while True:
    file_name = input("Enter the file name of the CSV or Excel file including Extension (.csv / .xlsx): ")
    print("You entered: ", file_name)
    if file_name.endswith('.csv'):
        df = pd.read_csv(file_name)
    elif file_name.endswith('.xlsx'):
        df = pd.read_excel(file_name)
    else:
        print("Unsupported format.\n Kindly upload CSV or Excel file only")
        continue
    if df is not None and not df.empty:
        print("\nFile data uploaded successfully. Processing...\n")
        print("\nThe columns are as follows: ", df.columns)
        print("Overview: ", df.head())
        break
    else:
        print("The file is empty. Retry.")
```

```python
# displaying list of columns
def show_columns():
    print("\n Current columns are:\n", df.columns.tolist())
    
    
#input.strip.lower.startswith()
def yes_or_no(prompt="Are you sure? (y/n): "):
    while True:
        m = input(prompt)
        if m.strip().lower().startswith('y'):
            return True
        elif m.strip().lower().startswith('n'):
            return False
        else:
            print("Invalid input.")
            print("Kindly enter y/n only")

            
#.isdigit()
def enter_no(min_val=1, max_val=10, prompt="Enter desired action [1-10]: "):
    while True:
        ch=input(prompt)
        if ch.isdigit():
            ch = int(ch)
            if min_val<= ch <= max_val:
                return ch
            else:
                print("INVALID Choice.")
                print("Kindly enter No. btw", min_val, "and", max_val)
        else:
            print("Kindly enter valid form above choices ONLY")


def col_no(min_val, max_val, prompt="Enter COLUMN NUMBER: "):
    while True:
        ah=input(prompt)
        if ah.isdigit():
                ah= int(ah)
                if min_val<= ah <= max_val:
                    return ah
                else:
                    print("OUT  OF BOUND")
                    print("Kindly enter No. btw", min_val, "and", max_val)

        else:
            print("Kindly enter valid form above choices ONLY")

def completed():
    print("\nProcess Completed\n")

    
#here df[col] means x is a list of column names and we are accessing index of that column list
x=df.columns.tolist() 


while True:
    print("A) HANDLING DATA\n")
    print("CHOOSE FROM GIVEN MENU TO DECIDE ACTION.\n Press [1-16] for desired output.\n")
    print("1. Check basic info of the dataset")
    print("2. Shape of Dataset  (Row,Columns)")
    print("3. Check datatype of each column")
    print("4. Column description")
    print("5. View Max, Min & Standard Deviation Values")
    print("6. View unique values in a column")
    print("7. Column missing data")
    print("8. Replace missing values with Mean/Median/Mode")
    print("9. Rename columns")
    print("10. Swap columns")
    print("11. Arrange columns alphabetically")
    print("12. Stats summary for numeric columns")
    print("13. View correlation matrrix btw Integers columns")
    print("14. View Combined Columns")
    print("15. Data Visualization")
    print("16. Exit Menu\n")
    while True:
        ch=enter_no(1,16,"Enter desired action [1-16]: ")
    
        #df.info(), df.shape
        if ch==1:
            df.info()
            print("\nShape of dataset: ", df.shape)
            completed()
    
        #df.shape
        elif ch==2:
            y=df.shape
            columns=int(y[1])
            rows= int(y[0])
            print("No.of rows :", rows)
            print("No.of columns :", columns)
            completed()
            
        #df.dtypes, df.value_counts()
        elif ch==3:
            print("Column wise data types: \n", df.dtypes)
            print("Count of each data type: \n", df.dtypes.value_counts())
            completed()
    
        #df[col].describe(),df.to_string()
        elif ch==4:
            while True:
                show_columns()
                ah =col_no(1, len(x), "Enter COLUMN NUMBER for description: ")
                s = df[x[ah - 1]].describe()
                print(s)
                
                if yes_or_no("Do you want to save the output as txt? (y/n): "):
                    y=input("Enter file name (without extension): ")
                    with open(y, "w") as f:
                        f.write(f" Summary for column: {x[ah - 1]}\n\n")
                        f.write(s.to_string())
                    print("Summary saved to",y,".txt")
                    completed()
                if not yes_or_no("Do you want to check description for another column? (y/n): "):
                    completed()
                    break
    
        #df[col].max,min,std()
        elif ch==5:
            int_col=[]
            non_int_col=[]
            for col in x:
                if df[col].dtype in ['int64','float64']:
                    int_col.append(col)
                else:
                    non_int_col.append(col)
    
            if not int_col:
                print("Numeric operations not possible on Non-Numeric data")
                completed()
    
            else:       
                print("Permissible columns:", int_col)    
                print("\nCHOOSE FROM GIVEN MENU TO VIEW: \n")
                print("1. Max Value")
                print("2. Min Value")
                print("3. Standard Deviation\n")
                
                while True:
                    bh=col_no(1,3,"Enter choice [1-3]: ")
                    print("Permissible columns:", int_col)
                    ah =col_no(1, len(x), "Enter COLUMN NUMBER: ")
                    
                    
                    if bh==1:
                        print("\nSELECTED CHOICE: Max Value\n") 
                        print("Max Value is: ", df[int_col[ah-1]].max())
                        completed()
                            
                    elif bh==2:
                        print("\nSELECTED CHOICE: Min Value\n")
                        print("Min Value is: ", df[int_col[ah-1]].min())
                        completed()
        
                    elif bh==3:
                        print("\nSELECTED CHOICE: Standard Deviation\n")
                        print("The standard deviation is:", df[int_col[ah-1]].std())
                        completed()
        
                    if not yes_or_no ("Do you want to calculate again? (y/n): "):
                        completed()
                        break
        
        #df[col].value_counts(dropna)
        elif ch==6:
            while True:
                show_columns()
                ah=col_no(1, len(x), "Enter COLUMN NUMBER for viewing unique values: ")
                
                if yes_or_no("Do you want to view NULLs too? (y/n): "):
                        print(df[x[ah-1]].value_counts(dropna=False))
                else:
                    print(df[x[ah-1]].value_counts(dropna=True))
        
                if not yes_or_no("Do you want to check for another column? (y/n): "):
                    completed()
                    break
    
        #df[col].isnull().sum()
        elif ch==7:
            while True:
                show_columns()
                ah =col_no(1, len(x), "Enter COLUMN NUMBER for viewing missing values: ")
                print("NULL VALUES FOUND: ", df[x[ah-1]].isnull().sum())
                
                if not yes_or_no("Do you want to check for any other column? (y/n): "):
                    completed()
                    break
    
        #df[col].fillna, df[col].mean/median/mode(), df[col].dtype
        elif ch==8:
            int_col=[]
            non_int_col=[]
            for col in x:
                if df[col].dtype in ['int64','float64']:
                    int_col.append(col)
                else:
                    non_int_col.append(col)
                    
            while True:
                show_columns()
                print("\nPERMISSIBLE COLUMNS for performing operations: ", int_col) 
                ah =col_no(1, len(int_col), "Enter COLUMN NUMBER for replacing null values: ")
        
                print("\nCHOOSE FROM GIVEN MENU TO REPLACE NULL DATA WITH: \n")
                print("1. Mean")
                print("2. Median")
                print("3. Mode\n")
                bh=col_no(1,3,"Enter choice [1-3]: ")
                
                if yes_or_no("Are you sure you want to replace? (y/n): "):
                    if bh==1:
                        print("The Mean is :", df[int_col[ah-1]].mean())
                        df[int_col[ah - 1]] = df[int_col[ah - 1]].fillna(df[int_col[ah - 1]].mean())
                                            
                    elif bh==2:
                        print("The Meadian is :", df[int_col[ah-1]].median())
                        df[int_col[ah - 1]] = df[int_col[ah - 1]].fillna(df[int_col[ah - 1]].median())
                            
                    elif bh==3:
                        print("The Mode is :", df[int_col[ah-1]].mode())
                        df[int_col[ah - 1]] = df[int_col[ah - 1]].fillna(df[int_col[ah - 1]].mode()[0])     
                        
                    if yes_or_no("Do you want to perform for another column? (y/n): "):
                        continue
                    else:
                        df.to_csv(file_name, index=False)
                        print("Changes saved to CSV file.")
                        completed()
                        break
                            
                else:
                    print("OK")
                    print("The Mean is :", df[int_col[ah-1]].mean())
                    print("The Median is :", df[int_col[ah-1]].median())
                    print("The Mode is :", df[int_col[ah-1]].mode()[0])
                        
                    if not yes_or_no("Do you want to perform in another column? (y/n): "):
                        completed()
                        break
    
        #df.rename(col={[i]:new_name},inplace)
        elif ch==9:
            while True:
                show_columns()
                ah =col_no(1, len(x), "Enter COLUMN NUMBER to be renamed: ")
                c = input("Enter NEW NAME for selected column: " + str(x[ah-1]))
                df.rename(columns={x[ah-1]:c}, inplace = True)
                df.to_csv(file_name, index=False)
                completed()
                if yes_or_no("Do you want to rename another column? (y/n): "):
                    continue
                else:
                    print("Column list updated in CSV file")
                    break
    
        #swap       
        elif ch == 10:
            x = df.columns.tolist()
            while True:
                show_columns()
                old_x = x.copy() 
        
                bh = col_no(1, len(x), "Enter COLUMN NUMBER to swap: ")
                ab = col_no(1, len(x), "Enter other COLUMN NUMBER to swap: ")
        
                x[bh-1], x[ab-1] = x[ab-1], x[bh-1] 
                df = df[x]
                print("\nColumns swapped\n")
                show_columns()
                
                if yes_or_no("Do you want to undo action? (y/n): "):
                    x = old_x.copy()
                    df = df[x]
                    print("Action withdrawn. Columns not swapped.")
                    show_columns()
                    break
                    
                else:
                    if yes_or_no("Do you want to save this change to CSV? (y/n): "):
                        df.to_csv(file_name, index=False)
                        print(f"Changes saved to '{file_name}'.")
                        completed()
                    else:
                        print("Changes not saved.")
                    if not yes_or_no("Do you want to swap more columns? (y/n): "):
                        completed()
                        break
        #sorting columns (asc or desc)
        elif ch==11:
            x = df.columns.tolist()
            while True:
                show_columns()
                old_x = x.copy()
                b= input("Press A/D for Ascending/Descending order resp. ")
                if b.strip().lower().startswith('a'):
                    sort_col=sorted(x)
                    df=df[sort_col] 
                    print("Columns sorted in ascending order.")
                    df.to_csv(file_name, index=False)
                elif b.strip().lower().startswith('d'):
                    sort_col=sorted(x, reverse=True)
                    df=df[sort_col] 
                    print("Columns sorted in descending order.")
                    df.to_csv(file_name, index=False)
                else:
                    print("Please RETRY with correct input.")
                    continue
                if yes_or_no("Do you want to undo the action? (y/n): "):
                    x = old_x.copy()
                    df = df[x]
                    print("Action withdrawn. Columns reverted to original order.")
                else:
                    print("Sorting confirmed.")
                    df.to_csv("your_filename.csv", index=False) 
                show_columns()
                completed()
                break
                  
        #df.describe()
        elif ch==12:
            show_columns()
            int_col=[]
            non_int_col=[]
            for col in x:
                if df[col].dtype == ['int64' ,'float64']:
                    int_col.append(col)
                else:
                    non_int_col.append(col)
    
            while True:
                print("Press A to view NUMERIC COLUMNS: ")
                print("Press B to view NON-NUMERIC COLUMNS: ")
                b=input('Enter choice: ')
                if b.lower().strip().startswith('a'):
                    print("NUMERIC COLUMNS:  ", int_col)
                    for i in int_col:
                        print(df[i].describe())
                    completed()
                    break
                elif b.lower().strip().startswith('b'):
                    print("NON-NUMERIC COLUMNS are: ", non_int_col)
                    print("Numeric stats Not Available for non- numeric columns.")
                    print("RETRY FOR NUMERIC COLUMNS")
    
                    if not yes_or_no("Do you want to proceed for the same? (y/n): "):
                        completed()
                        break
                else:
                    print("Please enter valid iput only.")
    
    
        #df[col].corr()
        elif ch==13:
            int_col = []
            non_int_col = []
            for col in x:
                if df[col].dtype in ['int64','float64']:
                    int_col.append(col)
                else:
                    non_int_col.append(col)
    
        
            if len(int_col) < 2:
                print("Not enough numeric columns for correlation.")
                completed()
            else:
                print("\nCorrelation Matrix (Numeric Columns Only):")
                corr = df[int_col].corr()
                print(corr)
    
            print("\nStrong Correlations (|value| â‰¥ 0.6): ")
            found = False
            for i in range(len(int_col)):
                for j in range(i + 1, len(int_col)):
                    col1 = int_col[i]
                    col2 = int_col[j]
                    value = corr.loc[col1, col2]
                    if abs(value) >= 0.6:
                        print(f"'{col1}' and '{col2}' have strong correlation: '{value}'")
                        found = True
            if not found:
                print("No strongly correlated column pairs found")
            completed()
    
    
    
        #combining cols
        elif ch==14:
            show_columns()
            completed_col=[]
            
            bh=enter_no(1,len(x),"\nEnter 1st column number for viewing combination: \n")
            completed_col.append(x[bh-1])
            show_columns()
            dh=enter_no(1,len(x),"\nEnter 2nd column number for viewing combination: \n")
            completed_col.append(x[dh-1])
            
            while yes_or_no("\nDo you want to combine more columns to view? (y/n): \n"):
                print("\nCombined Columns are: ", completed_col)
                show_columns()
                h=enter_no(1,len(x),"\nEnter column number (additional) for viewing combination: \n")
    
                if x[h-1] not in completed_col:
                    completed_col.append(x[h-1])
                    print("\nCombined Columns are: ", completed_col)
                else:
                    print("COLUMN ALREADY COMBINED\n")
                    print("Combined Columns are: ", completed_col)
    
                    if not yes_or_no("\nDo you want to add new column for combined view? (y/n): \n"):
                        break
    
            print("Combined columns are: ", df[completed_col])
            completed()
            
            #df[col].astype(). .apply(''), .join
            df['Combined Columns']=df[completed_col].astype(str).apply(' - '.join,axis=1)
            print(df['Combined Columns'])
                    
            if yes_or_no("Do you want to write it into a CSV file? (y/n): "):
                try:
                    df.to_csv(file_name, index=False)
                    print(f"'Combined Columns' added to '{file_name}'.")
                    
                    confirm_df = pd.read_csv(file_name)
                    print("Updated Columns:\n", confirm_df.columns.tolist())
                    completed()

                except Exception:
                    print("Error")
                    completed()
                df.to_csv(file_name, index=False)
                print(f"'Combined Columns' added to '{file_name}'.")
                completed()
            else:
                print("No file saved")
                completed()
                
        elif ch==15:
    
            print("B) VISUALIZE DATA\n")
            print("CHOOSE FROM GIVEN MENU TO DECIDE ACTION\n Press [1-5] for desired output\n")
            print("1. BAR GRAPH")
            print("2. HORIZONTAL BAR GRAPH")
            print("3. LINE GRAPH")
            print("4. PIE CHART")
            e=enter_no(1,5,"\nKindly enter choice from list: ")
            if e==1:
                while True:
                    print("\nYour choice: BAR GRAPH: ")
                    print("Press A for Default Bar Graph.")
                    print("\nPress B for Customized Bar Graph.")
                    a = input("\nEnter choice: \n")
                
                    if a.lower().strip().startswith('a'):
                        print("YOUR CHOICE: DEFAULT BAR GRAPH")
                        col = x[0]
    
                    elif a.lower().strip().startswith('b'):
                        print("YOUR CHOICE: CUSTOMIZED BAR GRAPH")
                        show_columns()
                        l = enter_no(1, len(x), "Enter COLUMN NUMBER: ")
                        col = x[l - 1]
    
                    else:
                        print("KINDLY ENTER VALID RESPONSE.")
                        continue
    
                    df[col].plot(kind='bar')
                    plt.xlabel('CATEGORIES')
                    plt.ylabel('FREQUENCY')
                    plt.title(f"Bar Graph for {col}")
                    plt.tight_layout()
                    plt.show()
                    completed()
                    break
                
            if e==2:
                while True:
                    print("\nYour choice: HORIZONTAL BAR GRAPH: ")
                    print("Press A for Default Horizontal Bar Graph.")
                    print("\nPress B for Customized Horizontal Bar Graph.")
                    a = input("\nEnter choice: \n")
                
                    if a.lower().strip().startswith('a'):
                        print("YOUR CHOICE: DEFAULT HORIZONTAL BAR GRAPH")
                        col = x[0]
    
                    elif a.lower().strip().startswith('b'):
                        print("YOUR CHOICE: CUSTOMIZED BAR GRAPH")
                        show_columns()
                        l = enter_no(1, len(x), "Enter COLUMN NUMBER: ")
                        col = x[l - 1]
    
                    else:
                        print("KINDLY ENTER VALID RESPONSE.")
                        continue
    
                    df[col].plot(kind='barh')
                    plt.xlabel('CATEGORIES')
                    plt.ylabel('FREQUENCY')
                    plt.title(f"Bar Graph for {col}")
                    plt.tight_layout()
                    plt.show()
                    completed()
                    break
                
            if e==3:
                while True:
                    print("\nYour choice: LINE GRAPH: ")
                    print("Press A for Default Line Graph.")
                    print("\nPress B for Customized Line Graph.")
                    a = input("\nEnter choice: \n")
                
                    if a.lower().strip().startswith('a'):
                        print("YOUR CHOICE: DEFAULT LINE GRAPH")
                        col = x[0]
    
                    elif a.lower().strip().startswith('b'):
                        print("YOUR CHOICE: CUSTOMIZED LINE GRAPH\n")
                        show_columns()
                        l = enter_no(1, len(x), "Enter COLUMN NUMBER: ")
                        col = x[l - 1]
    
                    else:
                        print("KINDLY ENTER VALID RESPONSE.")
                        continue
    
                    df[col].plot(kind='line')
                    plt.xlabel('CATEGORIES')
                    plt.ylabel('FREQUENCY')
                    plt.title(f"Line Graph for {col}")
                    plt.tight_layout()
                    plt.show()
                    completed()
                    break
                    
            elif e==4:
                while True:
                    print("Your choice: PIE-CHART: ")
                    show_columns()
                    l = enter_no(1, len(x), "Enter COLUMN NUMBER: ")
                    col = x[l - 1]
                    df[col].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
                    plt.title(file_name)
                    plt.ylabel('')
                    plt.tight_layout()
                    plt.show()
                    completed()
                    break
                                                
        elif ch==16:
            print("\nBye Bye...\n")
            completed()
            break
```
